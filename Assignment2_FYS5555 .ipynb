Discussion of the production and decay of the Higgs Boson in proton-proton collisions at the LHC:

Production of a Higgs boson in proton-proton collisions is extremely rare[1]. In fact, one needs to collide protons 40 million times per second to collect enough data to study the Higgs boson. After the production of the Higgs boson, we experience that the boson almost immediately decays into other elementary particles. The method to study the Higgs boson is therefore indirectly, and through a study of it's decay products. However, by increasing the TeV energy of the collisions, we increase the probability for observing the Higgs boson by many times[1]. 

There are four common ways to produce Higgs bosons in proton-proton colliders: vector-boson fusion (VBF), gluon fusion, associated production with a vector boson and associated production with a pair of top quarks. The production and decay modes are: H &rarr;$\gamma$$\gamma$, H&rarr;ZZ*&rarr;4l, H&rarr;WW*, H&rarr; $\tau$$\tau$ decay channels, H&rarr; $b$$\overline{b}$ decay channel probed in three exclusive production modes: in association with a weak vector boson (VH where V = W or Z) and in the H &rarr; $\mu$$\mu$ decay channel and H &rarr;Z$\gamma$ decay channel and more not mentioned here[2]. 

Though there are many decay particles, we know that a Higgs boson with a mass of 126 GeV has five main decay channels[3]. These decay channels are H to diphotons, ZZ, WW, bb and $\tau$$\tau$. It is these five processes we will continue on to make Feynman diagrams of. For the actual process of producing the particles the LHC accelerates two counter-rotating proton beams to nearly the speed of light so that the energy upon their collision should be sufficient to produce Higgs bosons over their expected mass range[3]. 

We call the set of digitized signals recording the collision products of a single crossing of proton packets an event. To quote the Atlas collaboration and their presentation of the finding of the Higgs boson: "a three-level trigger
system decides which events should be recorded; typically, 20 potentially interesting events are selected out of 1 million produced"[3]. The intention with this last method is to only record the most relevant events. The events deemed most interesting are those with high transverse momentum electrons, muons, photons, or jets[3]. We begin with explaning the phenomen known as on shell and off shell. The term decay width is well known for us from the curriculum. It is a fundamental variable to consider for all unstable particles. We have that the shorter the lifetime, the larger the width for the particle. It is usually determined from the invariant mass of its decay products, which takes the shape of a relativistic BreitWigner distribution[5]. For this we refer to the figure of the Breit-Wigner distribution at the end of this paragraph. If a Higgs boson is produced at the peak of 125 Gev in the plot, then it is on shell (this is the most probably way of producing Higgs boson). While it also possible for a production at high masses (off-shell) away from the peak. Higgs bosons produced on-shell depends not only on the Higgs bosons interactions with other particles, but also on the Higgs-boson width. To compare with this we see that the off-shell signal rate is independent, to a good approximation, of the Higgs-boson total width[5]. 


Some research has already been conducted on the process in (1). The ATLAS collaboration published such a report in 2023[6]. And a group led by Paul Avery et al. did a precision study with MEKD in 2012[7]. The Higgs boson to ZZ to 4 lepton channel is called the "golden channel" and has been widely used for studies on the Higgs boson due the fact that it offers the opportunity for clean measurements of the mass, spin, parity, etc. of the new resonance in a controlled environment
with low backgrounds[7]. 

Furthermore, the four lepton final state allows experiments to probe
the polarization of the intermediate Z bosons through angular correlations. The most used kinematic of the channel is the invariant mass. This is due to the large signal to background ratio for the four lepton mass[7]. When both Z boson are on shell, the interference effects between the leptons play an insignificant role. However, the inclusion of these interference effects plays an important role in achieving the best possible sensitivities in the low mass range and is
very important for separating spin zero and spin two resonance hypotheses.

```python
# cut on lepton charge
# paper: "selecting two pairs of isolated leptons, each of which is comprised of two leptons with the same flavour and opposite charge"
def cut_lep_charge(lep_charge):
# throw away when sum of lepton charges is not equal to 0
# first lepton in each event is [:, 0], 2nd lepton is [:, 1] etc
    return lep_charge[:, 0] + lep_charge[:, 1] + lep_charge[:, 2] + lep_charge[:, 3] != 0

# cut on lepton type
# paper: "selecting two pairs of isolated leptons, each of which is comprised of two leptons with the same flavour and opposite charge"
def cut_lep_type(lep_type):
# for an electron lep_type is 11
# for a muon lep_type is 13
# throw away when none of eeee, mumumumu, eemumu
    sum_lep_type = lep_type[:, 0] + lep_type[:, 1] + lep_type[:, 2] + lep_type[:, 3]
    return (sum_lep_type != 44) & (sum_lep_type != 48) & (sum_lep_type != 52)
```

```python
def calc_mllll(lep_pt, lep_eta, lep_phi, lep_E):
    # construct awkward 4-vector array
    p4 = vector.zip({"pt": lep_pt, "eta": lep_eta, "phi": lep_phi, "E": lep_E})
    # calculate invariant mass of first 4 leptons
    # [:, i] selects the i-th lepton in each event
    # .M calculates the invariant mass
    return (p4[:, 0] + p4[:, 1] + p4[:, 2] + p4[:, 3]).M

def calc_mll(lep_pt, lep_eta, lep_phi, lep_E):
    # construct awkward 4-vector array
    p4 = vector.zip({"pt": lep_pt, "eta": lep_eta, "phi": lep_phi, "E": lep_E})
    # calculate invariant mass of first 2 leptons
    # [:, i] selects the i-th lepton in each event
    # .M calculates the invariant mass
    return (p4[:, 0] + p4[:, 1]).M
```

```python
# Python method for computing MT2
# Uses the mt2 library
# Note this is vectorised...
def ComputeMT2(pts1,pts2,etas1,etas2,phis1,phis2,masses1,masses2,neutrino_pt,neutrino_phi):
    px1 = pts1 * np.cos(phis1)
    py1 = pts1 * np.sin(phis1)
    px2 = pts2 * np.cos(phis2)
    py2 = pts2 * np.sin(phis2)
    mpx = neutrino_pt * np.cos(neutrino_phi)
    mpy = neutrino_pt * np.sin(neutrino_phi)
    return mt2(masses1,px1,py1,masses2,px2,py2,mpx,mpy,0,0)
```

```python
# Function to build a record from the list of variable names
def buildRecord(awkwardEvents,variableDict):
    theDict = {}
    for key in variableDict.keys():
        if doDask: 
            theDict[key] = awkwardEvents[variableDict[key]]
        else:
            theDict[key] = awkwardEvents[variableDict[key]].array()
    return ak.zip(theDict,depth_limit=1)
```
